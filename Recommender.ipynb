### ***Third : Making a train & test set***

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(music_data, test_size=0.2, random_state=42)

print("Train set shape:", train_set.shape)
print("Test set shape:", test_set.shape)
```

```python
music_data = train_set.copy()
```

### ***Fourth : Feature normalization using Min-Max Scalar***

```python
from sklearn.preprocessing import MinMaxScaler

#numerical features for scaling
numerical_features = [
    'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'popularity' ,
    'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms'
]

# Initialize the scaler
scaler = MinMaxScaler()

# Normalize the numerical features
music_data[numerical_features] = scaler.fit_transform(music_data[numerical_features])

# Check the normalized values
music_data[numerical_features].head()
```

### ***Fifth : Defining the recommendation system***
#### ***Content-based recommendation system using cosine similarity***

```python
from sklearn.metrics.pairwise import cosine_similarity

# Function to recommend similar songs based on a song name
def recommend_similar_songs_by_name(music_data, song_name, top_n=10):
    # Ensure track_name column is included in music_data
    if 'track_name' not in music_data.columns:
        raise ValueError("DataFrame must contain 'track_name' column.")
    
    # Find the index of the song with the given name
    song_indices = music_data[music_data['track_name'].str.contains(song_name, case=False)].index
    if len(song_indices) == 0:
        raise ValueError(f"No song found with name '{song_name}'.")
    song_index = song_indices[0]
    
    # Get the features of the song
    song_features = music_data.loc[song_index, music_data.columns != 'track_name'].values.reshape(1, -1)
    
    # Calculate cosine similarity between the song and all other songs
    features = music_data.drop(columns=['track_name', 'genre', 'popularity'])  # Exclude non-numerical features for similarity calculation
    similarity_matrix = cosine_similarity(song_features, features)
    
    # Get the indices of the most similar songs
    sim_scores = list(enumerate(similarity_matrix[0]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Exclude the song itself and get top N similar songs
    similar_songs_indices = [i[0] for i in sim_scores if i[0] != song_index][:top_n]
    
    return music_data.iloc[similar_songs_indices]

# Test the function
similar_songs = recommend_similar_songs_by_name(music_data, 'your_favorite_song_name')
print(similar_songs[['track_name', 'artist_name', 'genre', 'popularity']])
```
